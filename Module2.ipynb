**Module 2: Data Visualization**

**#Principles of Effective Visualization-**

**Clarity**: The visualization should convey information clearly and accurately.

**Simplicity**: Avoid unnecessary complexity and clutter, focus on the essential message.
Relevance: Choose the right visualization type for the data and the story you want to tell.

**Aesthetics**: Make the visualization visually appealing and engaging.

**Context**: Provide enough context to help the viewer interpret the data.

**#Libraries-**

**Matplotlib**: The foundation for data visualization in Python. It provides a wide range of plotting functions.

**Seaborn**:
Built on top of Matplotlib, it simplifies the creation of attractive statistical visualizations.

**Pandas**:
Powerful library for data manipulation and analysis. It integrates seamlessly with Matplotlib and Seaborn.

**NumPy**:
Essential library for numerical operations in Python, often used in conjunction with Pandas.

**#Common Visualizations-**

**Bar plots:** Used to compare categorical data.
**Line plots:** Ideal for showing trends over time.
**Scatter plots:** Used to visualize relationships between two numerical variables.
**Histograms:** Display the distribution of a single numerical variable.
**Heatmaps:** Show the relationship between two variables using color intensity.
**Box plots:** Display the distribution and quartiles of a dataset.



```python
import pandas as pd
# Create a sample DataFrame
data = {
    'Name': ['Alice', 'Bob', 'Charlie', 'David'],
    'Age': [25, 30, 35, 40],
    'Salary': [50000, 60000, 70000, 80000]
}
df = pd.DataFrame(data)
```

```python
import matplotlib.pyplot as plt
import seaborn as sns

# Bar plot for salaries by name
plt.figure(figsize=(8, 5))
sns.barplot(x='Name', y='Salary', data=df)
plt.title('Salary by Name')
plt.xlabel('Name')
plt.ylabel('Salary')
plt.show()
```
